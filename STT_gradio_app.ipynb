```
import gradio as gr
import librosa
import numpy as np
import torch
from transformers import WhisperForConditionalGeneration, WhisperProcessor,AutoFeatureExtractor, Wav2Vec2BertModel,AutoProcessor,AutoModelForCTC,Wav2Vec2BertProcessor
```

```
model_w2v = AutoModelForCTC.from_pretrained("Cafet/w2v-bert-version-final").to("cuda")
processor_w2v = Wav2Vec2BertProcessor.from_pretrained("Cafet/w2v-bert-version-final")
processor_tiny = WhisperProcessor.from_pretrained("Cafet/whisper-tiny-final")
model_tiny = WhisperForConditionalGeneration.from_pretrained("Cafet/whisper-tiny-final").to("cuda")
processor_large = WhisperProcessor.from_pretrained("Cafet/whisper-large-final")
model_large = WhisperForConditionalGeneration.from_pretrained("Cafet/whisper-large-final").to("cuda")
```

```
def transcribe_w2v(audio):
    y, sr = librosa.load(audio)
    audio_resampled = librosa.resample(y, orig_sr=sr, target_sr=16000)
    input_id = processor_w2v(audio_resampled, sampling_rate=16000).input_features[0]
    input_features = torch.tensor(input_id).to("cuda")
    logits = model_w2v(input_features.unsqueeze(0)).logits
    pred_ids = torch.argmax(logits, dim=-1)[0]
    text = processor_w2v.decode(pred_ids)
    return text

def transcribe_large(audio):
    y, sr = librosa.load(audio)
    audio_resampled = librosa.resample(y, orig_sr=sr, target_sr=16000)
    input_id = processor_large(audio_resampled, sampling_rate=16000).input_features[0]
    input_id_3d = np.expand_dims(input_id, axis=0)
    inputs = torch.from_numpy(input_id_3d)
    generated_tokens = (
        model_large.generate(
            input_features=inputs.to("cuda"),
        ).cpu().numpy()
    )
    text = processor_large.tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    return text

def transcribe_tiny(audio):
    y, sr = librosa.load(audio)
    audio_resampled = librosa.resample(y, orig_sr=sr, target_sr=16000)
    input_id = processor_tiny(audio_resampled, sampling_rate=16000).input_features[0]
    input_id_3d = np.expand_dims(input_id, axis=0)
    inputs = torch.from_numpy(input_id_3d)
    generated_tokens = (
        model_tiny.generate(
            input_features=inputs.to("cuda"),
        ).cpu().numpy()
    )
    text = processor_tiny.tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    return text

def transcribe(audio, model_choice):
    if model_choice == "W2V bert Transcription":
        return transcribe_w2v(audio)
    elif model_choice == "Whisper Large Mongolian":
        return transcribe_large(audio)
    elif model_choice == "Whisper Tiny Mongolian":
        return transcribe_tiny(audio)

iface = gr.Interface(
    fn=transcribe,
    inputs=[
        gr.Audio(source="microphone", type="filepath"),
        gr.Dropdown(["W2V Mongolian Transcription", "Whisper Large Mongolian", "Whisper Tiny Mongolian"], label="Model")
    ],
    outputs="text",
    title="Mongolian Transcription Models /NVIDIA-T4-gpu/",
    description="Select a model to transcribe Mongolian speech."
)

iface.launch(share=True)
```
